## Breast Cancer Prediction using classification algorithms
### Decision Tree, Support Vector Machine and Random Forest 

```python
## Import libraries for the project

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

from sklearn import preprocessing
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import linear_model
from sklearn.svm import SVR
from scipy import stats

import os
```

```python
## Load the dataset
bc_data = pd.read_csv('C:\\Users\\user\\Desktop\\DataScience\\BreastCancer\\data.csv')
bc_data.head()
```

```python
# diagnosis column was replaced as the target variable in the dataframe

bc = load_breast_cancer(as_frame=True)
bc_df = pd.DataFrame(data=bc.data, columns=bc.feature_names)
bc_df['diagnosis'] = bc.target
bc_df.head(3)
```

```python
## introducing a heatmap to display the correlation of all features   
plt.figure(figsize=(20,20))
sns.heatmap(bc_df.corr(), annot=True)
```

from the correlation analyis the is negative correlation (>0.7) with some variable like "worst concave points", "worst area", "worst perimeter", "worst radius", "mean concave point", "mean concavity", "mean area", "mean perimeter" and "mean radius" with the breast cancer diagnosis

```python
# Change the value of "diagnose" in bc_df to 0: Malignant and 1: Benign
bc_df['diagnosis'] = bc_df.diagnosis.map({0:"Malignant",
                               1:"benign"})
bc_df['diagnosis'].value_counts()
```

```python
## comparing the values of the variable that is the most strongly correllated using box plot
sns.catplot(x="worst concave points", y="diagnosis", data=bc_df, kind="box", height=4, aspect=2)
```

### Splitting of the data for machine learning classification

```python
# Splitting into 80/20 train-test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bc_df.drop('diagnosis', axis=1),  ## drop the target variable for x axis
                                                   bc_df['diagnosis'],  ## use the target variable for y axis
                                                   stratify=bc_df['diagnosis'],  ## Stratify the split
                                                   test_size=0.2,
                                                   random_state=123)
```

```python
# Displaying the stratified data split for the training and splitting dataset
fig, ax=plt.subplots(1,2, figsize=(8,3))
sns.countplot(x=y_train, ax=ax[0], order=["Malignant","benign"])
sns.countplot(x=y_test, ax=ax[1], order=["Malignant","benign"])
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier()

# fit the tree to the training set
model_tree.fit(X_train, y_train)

print('Training accuracy: ', model_tree.score(X_train, y_train))
```

### Evaluation Metrics

```python
from sklearn.metrics import classification_report, plot_confusion_matrix

## using the fitted model_tree, predict the outcome from the test set
pred_tree = model_tree.predict(X_test)

## we need to print out the report using the imported metrics
# classification report
print(classification_report(y_test, pred_tree))

# Plot the confusion matrix
plot_confusion_matrix(model_tree, X_test, y_test)
```

The model was able to predict accuracy of 95% which is very good and very close to the 96.9% gold standard benchmark. The precision of Malignant is 91% and Benign is 99% right which is also good. On the other hand the recall is also very good as it was able to correctly classify the Malignant by 98% and benign by 94% which looks amazing.
The F1 which is the harmonic mean between the precision and recall is also great.

### Hyper-Parameter tuning
To see the effect of changing some of the parameters

```python
def tree_report(model):
    test_pred = model.predict(X_test)  #prediction of the test set
    
    # print the classification report
    print(classification_report(y_test, test_pred))
    
    #plot the confusion matrix
    plot_confusion_matrix(model, X_test, y_test)
    
    # plot the resulting tree
    plt.figure(figsize=(20,10), dpi=300)
    plot_tree(model, feature_names=X_train.columns);
```

```python
# Initiate a pruned tree with depth=3
model_pruned = DecisionTreeClassifier(criterion="gini",
                                      min_samples_split=2,
                                      min_samples_leaf=1,
                                      max_depth=4, 
                                      random_state=123)

#Fit the pruned tree to the training set
model_pruned.fit(X_train, y_train)

# Generate the report
tree_report(model_pruned)
```

After adjusting the settings of the hyper-parameters, the performance of the model slighly increased for the classification metrics

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=20,      # number of trees to 20
                                 random_state=123)

# Fit the model to the training set
rf_model.fit(X_train, y_train)

# Predict from the test set
rf_pred = rf_model.predict(X_test)

# classification report and confusion matrix
print(classification_report(y_test, rf_pred))
plot_confusion_matrix(rf_model, X_test, y_test)
```

Since this is an improved model for decision tree, the model was able to predict better with an accuracy of 99% which is above the 96.9% gold standard benchmark. The precision of Malignant is 98% and Benign is 100%. On the other hand the recall and F1 are also very good.

### Using Cross Validation for Random Forest

```python
## We use grid search to decide the two hyperparameters.

from sklearn.model_selection import GridSearchCV

n_estimators = [32, 64, 96, 128, 160]
max_features = [4,5,6,7]
param_grid = {'n_estimators': n_estimators,
              'max_features': max_features}

rf_model = RandomForestClassifier()
grid = GridSearchCV(rf_model, param_grid)
grid.fit(X_train, y_train)
```

```python
# best parameters
grid.best_params_
```

```python
# Fit the model using the best chosen parameters and report the performance
rfcv_model = RandomForestClassifier(max_features=5, n_estimators=128)
rfcv_model.fit(X_train, y_train)
rfcv_pred = rfcv_model.predict(X_test)
```

```python
print(classification_report(y_test, rfcv_pred))
plot_confusion_matrix(rfcv_model, X_test, y_test)
```

### Error vs. number of trees

Plot the errors and number of misclassifications against the number of trees.

```python
from sklearn.metrics import accuracy_score

errors = []
misclassifications = []

for n in range(1,200,5):
    model = RandomForestClassifier(max_features=4, n_estimators=n)
    model.fit(X_train, y_train)
    test_pred = model.predict(X_test)
    
    e = 1-accuracy_score(y_test, test_pred)
    missed = (y_test!=test_pred).sum() 
    
    errors.append(e)
    misclassifications.append(missed)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))
# Plot errors and misclassifications side by side
ax1.plot(range(1,200,5), errors)
ax2.plot(range(1,200,5), misclassifications)
```

The line exhibited an oscillatory trend. Instead of using the 128 (number of trees) as selected by the cross validation, we can decide to stop and use a fewer tree of around 40.
